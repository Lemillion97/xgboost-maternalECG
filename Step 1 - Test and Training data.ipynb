```python
# ================================
# Maternal Health Risk - Step A
# Train/Test Split (Paper-Aligned)
# ================================

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# -------------------------------
# 1. Load dataset
# -------------------------------
file_path = "C:/Users/vivek/Desktop/KI_Project/Project 2/MaternalDS.csv"  # update if needed
df = pd.read_csv(file_path)

print("Dataset loaded successfully.\n")
print(df.head())
```

```python
# -------------------------------
# 2. Convert target to binary
# -------------------------------
# high risk → 1 (positive class)
# low + mid risk → 0 (negative class)

df['RiskLevel'] = df['RiskLevel'].map({
    'low risk': 0,
    'mid risk': 0,
    'high risk': 1
})
print("\nTarget encoding complete:")
print(df['RiskLevel'].value_counts())
```

```python
# -------------------------------
# 3. Check class distribution
# -------------------------------
print("\nClass distribution (proportion):")
print(df['RiskLevel'].value_counts(normalize=True))

# -------------------------------
# 4. Define features and target
# -------------------------------
X = df.drop('RiskLevel', axis=1)
y = df['RiskLevel']
```

```python
# -------------------------------
# 5. Stratified train/test split
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
# -------------------------------
# 6. Verify split
# -------------------------------
print("\n=== Split Summary ===")
print(f"Train size: {X_train.shape}")
print(f"Test size: {X_test.shape}")

print("\nTrain class distribution:")
print(y_train.value_counts(normalize=True))

print("\nTest class distribution:")
print(y_test.value_counts(normalize=True))
```

```python
# -------------------------------
# 7. Sanity checks
# -------------------------------
assert len(X_train) + len(X_test) == len(df), "Split size mismatch!"
assert set(y_train.unique()) == set(y_test.unique()), "Class missing in split!"

print("\n✅ Train/Test split completed successfully (no leakage, stratified, reproducible).")
```

```python
# -------------------------------
# 8. Saving the files to computer
# -------------------------------

X_train.to_csv("C:/Users/vivek/Desktop/KI_Project/Project 2/X_training.csv", index=False)
X_test.to_csv("C:/Users/vivek/Desktop/KI_Project/Project 2/X_test.csv", index=False)
y_train.to_csv("C:/Users/vivek/Desktop/KI_Project/Project 2/Y_training.csv", index=False)
y_test.to_csv("C:/Users/vivek/Desktop/KI_Project/Project 2/Y_test.csv", index=False)
```
